# What is the difference between interpreted and compiled languages?
"Interpreted languages (like Python) execute code line by line using an interpreter, while compiled languages (like C++) convert the entire code to machine code before running. Interpreted code runs directly, but compiled code needs to be built first."

# What is exception handling in Python?
"Exception handling lets you manage runtime errors using try, except, else, and finally blocks. It prevents program crashes and lets you handle problems gracefully."

# What is the purpose of the finally block in exception handling?
"The finally block always runs, whether or not an exception occurred. It is commonly used to clean up resources like closing files or releasing memory."

# What is logging in Python?
"Logging is used to record messages, errors, or events that happen during a program’s execution. It's useful for debugging and monitoring code without using print statements."

# What is the significance of the __del__ method in Python?
"__del__ is a special method called when an object is about to be destroyed. It is used to perform clean-up actions, like closing files or freeing resources."

# What is the difference between import and from ... import in Python?
"'import module' loads the entire module (you access functions like module.func()). 'from module import func' imports just the function, so you can use it directly as func()."

# How can you handle multiple exceptions in Python?
"Use multiple except blocks for different errors:\n\ntry:\n    x = int('abc')\nexcept ValueError:\n    print('ValueError caught')\nexcept TypeError:\n    print('TypeError caught')"

# What is the purpose of the with statement when handling files in Python?
"The with statement automatically opens and closes files. It ensures that the file is properly closed after reading or writing, even if an error occurs."

# What is the difference between multithreading and multiprocessing?
"Multithreading runs multiple threads in the same process (good for I/O tasks), while multiprocessing runs separate processes (good for CPU tasks). Threads share memory, processes do not."

# What are the advantages of using logging in a program?
"Logging helps you debug, monitor, and trace code without using print statements. It records errors and messages to a file or console for future analysis."

# What is memory management in Python?
"Memory management means allocating and releasing memory as needed. Python handles this automatically using garbage collection and reference counting."

# What are the basic steps involved in exception handling in Python?
"1. Wrap risky code in a try block\n2. Handle exceptions in the except block\n3. Optionally use else if no error occurs\n4. Use finally to run code regardless of errors"

# Why is memory management important in Python?
"Good memory management prevents memory leaks, ensures efficient use of resources, and improves program performance. It is important in large or long-running apps."

# What is the role of try and except in exception handling?
"The try block contains code that might raise an error. The except block catches and handles the error without crashing the program."

# How does Python's garbage collection system work?
"Python uses reference counting and a garbage collector to automatically delete unused objects and free up memory."

# What is the purpose of the else block in exception handling?
"The else block runs only if no exception occurs in the try block. It is placed after all except blocks."

# What are the common logging levels in Python?
"1. DEBUG – Detailed information for debugging\n2. INFO – General program status\n3. WARNING – Something unexpected\n4. ERROR – Serious error\n5. CRITICAL – Very serious error"

# What is the difference between os.fork() and multiprocessing in Python?
"os.fork() is Unix-only and creates a child process. multiprocessing is cross-platform and provides a safer, easier way to create and manage multiple processes."

# What is the importance of closing a file in Python?
"Closing a file frees system resources and ensures that data is written properly. Using with statements handles closing automatically."

# What is the difference between file.read() and file.readline() in Python?
"file.read() reads the entire file at once. file.readline() reads one line at a time, which is useful for large files."

# What is the logging module in Python used for?
"The logging module lets you record events, messages, and errors in your program for debugging, auditing, or monitoring."

# What is the os module in Python used for in file handling?
"The os module lets you interact with the file system: check if files exist, get file size, create/delete folders, and more."

# What are the challenges associated with memory management in Python?
"Challenges include circular references, large objects staying in memory, and memory leaks in C extensions. Managing these helps performance."

# How do you raise an exception manually in Python?
"Use the raise keyword:\n\nraise ValueError('This is a custom error')"

# Why is it important to use multithreading in certain applications?
"Multithreading allows your program to handle multiple tasks at once (like downloading files or handling users), making it faster and more efficient for I/O tasks."

```python
#How would you implement a basic logging setup that logs to a file with rotation after 1MB
import logging
from logging.handlers import RotatingFileHandler

logging.basicConfig(level=logging.INFO)
handler = RotatingFileHandler("app.log", maxBytes=1000000, backupCount=5)
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)
logging.getLogger().addHandler(handler)
logging.info("This is an info message.")
```
